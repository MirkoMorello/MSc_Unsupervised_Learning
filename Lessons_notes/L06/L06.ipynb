# Cluster Analysis: K-means Clustering

### Partitional Clustering approach
We have a number of cluster, K, to be specified, then each cluster is associated with a centroid (center point), each point is asigned to the cluster with the closest centroid, the basic algorithm is very simple:
```
Select K points as the initial centroids  
DO:
  
  Form K clusters by clusters by assigning all points to the closest centroid
  recompute the centroid of each cluster

WHILE the centroids don't change
```

last iteration is missing but would have been the same as the last panel, and the algorithm would have stopped

### Common objective function: Sum of Squared Error (SSE)
$$SSE = \sum_{i=1}^{K} \sum_{x \isin C_i} dist^2(m_j,x) = \sum_{i=1}^{K} \sum_{x \isin C_i}\sum_{j=1}^{n}(m_{ij} - x_{ij})^2$$

### Choosing the initial centers:
there can be problems with choosing the initial centers, if there are K 'real clusters then the chance of selecting one centroid from each cluster is small.
