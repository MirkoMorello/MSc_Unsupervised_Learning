Generally, we can regard Cluster analysis as, Given a set of objects, place them in groups such that the objects in a group are similar (or related) to one another and different from (or unrelated to) the objects in other groups.

I want both to minimize the intra-cluster distances and maximize the inter-cluster distances.

## Types of Clustering:

A CLustering is a set of clusters, important distinction between partitional and hierarchical sets of clusters:
- **Partitional Clustering**: a division od fata objects into non-overlapping subsets
- **Hierarchical Clustering**: a set of nester clusters organized as a hierarchical tree (matrioska)

We can also make a distiction between sets of clusters
- **Exclusive** versus **non-exclusive**
  - In non exclusive clusterings, points may belong to multiple clusters
  - can belong to multiple classes (clusters) or should be 'Border' points
- **Fuzzy clustering**
  - In fuzzy clustering, a point belongs to every cluster with some weight between 0 and 1
  - weights must sum to 1
  - probabilistic clustering has similar chracteristics
- **Partial** versus **complete**
  - in some cases, we only want cluster some of the data
  - data c an contain outliser or anomalous observations

- Well-separated clusters, intra-distance is always less than inter-distance
- Not Well-separated clusters, intra distance may be more than inter-distance
- Prototype-based clusters, each objects is closer (more similar) to the protoype that defines the cluster than to the prototype of any other cluster. For continuous attributes, the protoype of a cluster is often the **centroid**, the average of all the objects in the cluster, the intra-distance from the prototype is always less than the distance to the other cluster's prototype.
- Contiguity-based clusters, it's a set of point such tant a point in the cluster is closer (or more similar) to one or more other points in the cluster tan to any point not in the cluster. They can assume values like a banana
- Density-based clusters, The observations of a given cluster are such taht they define a high density area of the space, which is alternated to a low density area, and then again alternated to a high density area.
- Described by an objective Function, finds clusters that minimize or maximize an objective function
  - Enumerate all possible ways of dividing the point into clusters and evaluate thte 'goodness' of each potential set of clusters by using the given objective function (NP hard)
