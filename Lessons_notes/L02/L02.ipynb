## Data object and attribute:
Data is the fundamental component of measuring a system.  
It's a collentcion of **data objects** abd they **attributes**, an attribute is a properrty or characteristic of an object  
A collection of attributes desribe an **object** (record, point, case, entity, instance are all synonims)  
**Attribute values** are numbers or symbols assigned to an attribute for a particular object, an ettribute, like height, can be mapped to different attribute values due to the fact there are multiple way to measure it, like feet and meters.  
Type of attributes:
- Nominal (like id, eye color, zip codes etc.)
- Ordinal (like ranking, grades)
- Interval (calendar dates, temperatures)
- Ratio

The type of ana attribute depends on which of the following properties/operations t possesses:
- Distintness = !=
- Order < >
- Differences are meaningful  - +
- Ratios are meaningful * /

**Asymmetric attribute:** When we have a lot of zeroes and we have the presence of a non-zero attribute value, we don't care about thea actual value, but the only presence alone is important.
